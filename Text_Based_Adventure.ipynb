```python
def intro():
  response = input("Do you want to learn about the fish or the sharks first? Fish or Sharks: ")
  return response
```

```python
def sharkRoom():
  response = input("You have now entered the shark exibate, do you want to learn about the Thresher shark or the Leopard Shark? Thresher or Leopard: ")
  return response
```

```python
def fishRoom():
  response = input("You have enter the fish exibate, would you like to learn about the Mahi-Mahi fish or the school of Clownfish? Mahi-Mahi or Clowfish: ")
  return response
```

```python
def main():
  response = intro()
  if response =="Fish":
    response = fishRoom
  elif response=="Sharks":
    response = sharkRoom()
```
